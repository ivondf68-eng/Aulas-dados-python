```python
# Importando biblioteca pandas para manipulação de dados
import pandas as pd
```

```python
# Lê o arquivo CSV de uma URL e o carrega em um DataFrame Pandas
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
# Verifica a presença de valores nulos em cada célula do DataFrame
df.isnull()
```

```python
# Define um dicionário para mapear os nomes das colunas antigas para novos nomes em português
column_mapping = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
# Renomeia as colunas do DataFrame usando o dicionário de mapeamento
df = df.rename(columns=column_mapping)
```

```python
# Exibe as primeiras linhas do DataFrame
df.head()
```

```python
# Calcula a soma de valores nulos por coluna no DataFrame
df.isnull().sum()
```

```python
# Exibe os valores únicos da coluna 'ano'
df["ano"].unique()
```

```python
# Filtra o DataFrame para exibir apenas as linhas que contêm pelo menos um valor nulo
df[df.isnull().any(axis=1)]
```

```python
# Importando a biblioteca NumPy
import numpy as np

#criação de um Dataframe
df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'José', 'Ana', 'Val'],
    'salario': [5000, np.nan, 7000, np.nan, 100000]
})

#calcula a média, substitui os números nulos e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#calcula a mediana e substitui os números nulos
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

print(df_salarios)
```

```python
# DataFrame de exemplo com dias da semana e temperaturas, incluindo valores nulos
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenche os valores nulos na coluna 'Temperatura' com o último valor válido observado (forward-fill)
df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
# Cria um DataFrame de exemplo com dias da semana e temperaturas
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

# Preenche os valores nulos na coluna 'Temperatura' com o próximo valor válido observado (backward-fill)
df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
# Cria um DataFrame de exemplo com nomes e cidades, incluindo valores nulos
df_cidades = pd.DataFrame({
    'nome': ['João', 'Maria', 'José', 'Ana', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Belo Horizonte', np.nan, 'Rio de Janeiro']
})

# Preenche os valores nulos na coluna 'cidade' com a string 'Não informado'
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```
