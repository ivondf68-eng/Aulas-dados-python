```python
#Importando biblioteca pandas para manipulação de dados
import pandas as pd
```

```python
#Carrega o arquivo csv de uma URL
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
# Exibe as primeiras 10 linhas do DataFrame para uma visualização inicial dos dados.
df.head(10)
```

```python
# Exibe um resumo do DataFrame, incluindo o tipo de dados e o número de valores não nulos em cada coluna.
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
linhas, colunas = df.shape[0], df.shape[1]
print("linhas: " , linhas)
print("colunas: " , colunas)
```

```python
# Renomeia as colunas do DataFrame para termos em português brasileiro, facilitando a compreensão e manipulação dos dados.
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})

print("Colunas traduzidas com sucesso!")
print(df.columns)
```

```python
# Conta a ocorrência de cada nível de experiência na coluna 'senioridade'.
df["senioridade"].value_counts()
```

```python
# Conta a ocorrência de cada tipo de contrato na coluna 'contrato'.
df["contrato"].value_counts()
```

```python
# Conta a ocorrência de cada modalidade de trabalho (presencial, remoto, híbrido) na coluna 'remoto'.
df["remoto"].value_counts()
```

```python
# Conta a ocorrência de cada tamanho de empresa na coluna 'tamanho_empresa'.
df["tamanho_empresa"].value_counts()
```

```python
# Traduz as categorias da coluna 'senioridade' de códigos para termos em português
df['senioridade'] = df['senioridade'].map({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
print("Categorias da coluna 'senioridade' traduzidas com sucesso!")
print(df['senioridade'].value_counts())
```

```python
# Traduz as categorias da coluna 'contrato' de códigos para termos em português.
df['contrato'] = df['contrato'].map({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelance'
})
print("Categorias da coluna 'contrato' traduzidas com sucesso!")
print(df['contrato'].value_counts())
```

```python
# Traduz as categorias da coluna 'tamanho_empresa' de códigos para termos em português.
df['tamanho_empresa'] = df['tamanho_empresa'].map({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
print("Categorias da coluna 'tamanho_empresa' traduzidas com sucesso!")
print(df['tamanho_empresa'].value_counts())
```

```python
# Traduz as categorias da coluna 'remoto' de códigos numéricos para termos em português.
df['remoto'] = df['remoto'].map({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
print("Categorias da coluna 'remoto' traduzidas com sucesso!")
print(df['remoto'].value_counts())
```
